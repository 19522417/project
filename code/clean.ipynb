```python
  #-- 3. chest pain type  (4 vales)       
  ## 9 cp: chest pain type
  ##      -- Value 1: typical angina
  ##      -- Value 2: atypical angina
  ##      -- Value 3: non-anginal pain
  ##      -- Value 4: asymptomatic
id=[1,2,3,4]
cp_value=[
      "typical angina",
      "atypical angina",
      "non-anginal pain",
      "asymptomatic"
]
data_cp=pd.DataFrame()
data_cp["id"]=id
data_cp["cp"]=cp_value
data_cp.to_csv('../data/data_cp.csv',index=False)
data_cp.head()
```

```python
#16 fbs: (fasting blood sugar > 120 mg/dl)  (1 = true; 0 = false)
 
id=[1,0]
fbs_value=[
    True,
    False
]
data_fbs=pd.DataFrame()
data_fbs["id"]=id
data_fbs["fbs"]=fbs_value
data_fbs.to_csv('../data/data_fbs.csv',index=False)
data_fbs.head()
```

```python
  #-- 7. resting electrocardiographic results  (values 0,1,2) 
 # 19 restecg: resting electrocardiographic results
 #       -- Value 0: normal
 #       -- Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
 #       -- Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
id=[0,1,2]
restecg_value=["normal", " having ST-T wave abnormality","Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria"]
data_restecg=pd.DataFrame()
data_restecg["id"]=id
data_restecg["restecg"]=restecg_value
data_restecg.to_csv('../data/data_restecg.csv',index=False)
data_restecg.head()
```

```python
 # exang: exercise induced angina (1 = yes; 0 = no)


id=[0,1]
exang_value=[
     "yes",
     "no"
]
data_exang=pd.DataFrame()
data_exang["id"]=id
data_exang["exang"]=exang_value
data_exang.to_csv('../data/data_exang.csv',index=False)
data_exang.head()  
```

```python
 ## 41 slope: the slope of the peak exercise ST segment
 ##      -- Value 1: upsloping
 ##      -- Value 2: flat
 ##      -- Value 3: downsloping
id=[1,2,3]
slope_value=[
      "upsloping",
      "flat",
      "downsloping"
]
data_slope=pd.DataFrame()
data_slope["id"]=id
data_slope["slope"]=slope_value
data_slope.to_csv('../data/data_slope.csv',index=False)
data_slope.head()  
```
